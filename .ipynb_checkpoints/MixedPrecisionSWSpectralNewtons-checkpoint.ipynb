```julia
#9/4/2024

using FFTW
using LinearAlgebra
using MultiFloats
MultiFloats.use_bigfloat_transcendentals()
using Plots
using BenchmarkTools
using Printf

flush(stdout)

function fourier_diff_matrix_naive(N::Int; L::Real = 2π, T = Float64)
    @assert iseven(N) "N must be even."

    k = vcat(0:N÷2, -N÷2+1:-1)
    k = T.(2π / L .* k)
    D1_operator = Complex{T}.(im .* k)

    ω = exp(-2π * im / N)
    F = [ω^((m-1)*(n-1)) for m in 1:N, n in 1:N] ./ sqrt(T.(N))
    F_inv = inv(F)

    D = real(F_inv * Diagonal(D1_operator) * F)
    return Matrix{T}(D)  # force type T here
end


function newtons_method(f, x0, Dx, dt, tol=1e-10, maxiter=100)
    x = x0
    T = typeof(x0[1])
    g = T(9.81)
    N = length(x0) ÷ 2
    Id = Matrix{T}(I, length(x0), length(x0))  
    for i in 1:maxiter
        fx = f(x)
        if norm(fx) < tol
            return x
        end
        x1 = x[1:N]
        x2 = x[N+1:end]
        A11 = zeros(eltype(x1), N, N)
        A12 = Dx
        A21 = -Dx * Diagonal((x2.^2) ./ (x1.^2) .- g .* x1)
        A22 = 2 * Dx * Diagonal(x2 ./ x1)
        A = [A11  A12; A21  A22]
        J = Id + (dt / 2) * A
        x = x - (J \ fx)
    end
    return x
end

function implicit_midpoint_step(
    u::Vector{H}, D::Matrix{H}, Dlow::Matrix{L}, dt::H,
    ::Type{H}, ::Type{L};
    correct::Bool=false, n_corr::Int=2,
    B::Matrix{H}, update::Bool=true
) where {H, L}

    ulow = L.(u)
    N = length(u) ÷ 2  
    g = L(9.81)

    f = y -> y .- ulow .+ (L(dt)/2) .* (
        [ Dlow * y[N+1:end];
          Dlow * ((y[N+1:end].^2) ./ y[1:N] .+ (g/2) .* (y[1:N].^2)) ]
    )

    tol = 100 * eps(L)
    y1 = newtons_method(f, ulow, Dlow, L(dt), tol)
    y1 = H.(y1)


    
    if correct
        tol_corr = 100*eps(H)
        g = H(9.81)
        f = y -> y .- u .+ (dt/2) .* (
            [ D * y[N+1:end];
              D * ((y[N+1:end].^2) ./ y[1:N] .+ (g/2) .* (y[1:N].^2)) ]
        )

        y0 = copy(y1)
        r0 = f(y0)
        r1 = copy(r0)
        dy = similar(y0)
        df = similar(y0)
        E  = similar(B)
        Id = Matrix{H}(I, 2*N, 2*N)

        for i in 1:n_corr
            y1 .= y0 .- B*r0
            if update
                r1 .= f(y1)
                if norm(r1) < tol_corr
                    break
                end
                dy .= y1 .- y0          # broadcast avoids temporary array for subtraction
                df .= r1 .- r0
    
                denom = dot(df, df)    # same as (b * df), scalar
                if denom != 0
                    E .= (dy .- B*df) * df' / denom   # df' is the adjoint (row vector), no extra b
                    B .+= E                          # in-place update, avoids new matrix
                end
            end
            y0 .= y1               # copy values instead of rebinding
            r0 .= r1
        end    
    end
    return u - (dt) .* (
        [ D * y1[N+1:end];
          D * ((y1[N+1:end].^2) ./ y1[1:N] .+ (g/2) .* (y1[1:N].^2)) ]
    )
end
function rk4_step(u, D, dt)
    n = length(u) ÷ 2
    g = 9.81
    top1 = -D * u[n+1:end]
    bot1 = -D * ((u[n+1:end].^2) ./ u[1:n] .+ (g/2) .* (u[1:n].^2))
    k1 = [top1; bot1]

    v = u .+ 0.5 * dt .* k1

    top2 = -D * v[n+1:end]
    bot2 = -D * ((v[n+1:end].^2) ./ v[1:n] .+ (g/2) .* (v[1:n].^2))
    k2 = [top2; bot2]

    v = u .+ 0.5 * dt .* k2

    top3 = -D * v[n+1:end]
    bot3 = -D * ((v[n+1:end].^2) ./ v[1:n] .+ (g/2) .* (v[1:n].^2))
    k3 = [top3; bot3]

    v = u .+ dt .* k3

    top4 = -D * v[n+1:end]
    bot4 = -D * ((v[n+1:end].^2) ./ v[1:n] .+ (g/2) .* (v[1:n].^2))
    k4 = [top4; bot4]

    return u .+ (dt/6) .* (k1 .+ 2 .* k2 .+ 2 .* k3 .+ k4)
end

function norm_L2(u, dx)
    return sqrt(sum(abs2, u) * dx)
end


using Dates  # for timing

function run_mixed_precision_burgers(low::Type, high::Type; N=50, T_Final=0.3, correct::Bool= false, corr_id::Int = 0, n_corr = 1)
    L = 2π
    x = L * (0:N-1) / N
    dx = L / N


    # Use naive diff for high, FFTW for low as long as low is fp64 or less
    Dx_high = fourier_diff_matrix_naive(N; L=L, T=high)
    Dx_low  = fourier_diff_matrix_naive(N; L=L, T=low)
    #Dx_high = fourier_diff_matrix(N; L=L, T=high)  # FFTW for high precision
    #Dx_low  = fourier_diff_matrix(N; L=L, T=low)   # FFTW for low precision

    
    u0 = zeros(2N)
    u0[1:N] .= high.(cos.(BigFloat.(x)) .+ 2)   # h = cos(x)+2
    u0[N+1:end] .= zero(high)                   # v = 0

    # Reference RK4 solution
    dt_rk = 1e-4
    nt_rk = Int(round(T_Final / dt_rk))
    uref = copy(u0)

    for _ in 1:nt_rk
        uref = rk4_step(uref, Dx_high, dt_rk)
    end

    # Run mixed-precision IMR
    dt = 1e-3
    nt = Int(round(T_Final / dt))
    u = copy(u0)

    Id_high = Matrix{high}(I, 2*N, 2*N)
    if correct 
        if corr_id == 1
            B = Id_high
            upd = false
        elseif corr_id == 2
            Id = Matrix{low}(I, 2*N, 2*N)
            
            x1 = low.(u0[1:N])        # convert to low precision
            x2 = low.(u0[N+1:end])    # convert to low precision
            gL = low(9.81)
            
            A11 = zeros(low, N, N)
            A12 = Dx_low                          # already low precision
            A21 = -Dx_low * Diagonal((x2.^2) ./ (x1.^2) .- gL .* x1)
            A22 = 2 * Dx_low * Diagonal(x2 ./ x1)
            
            A = [A11  A12;
                 A21  A22]
            
            J = Id + (low(dt) / 2) * A   # now J is fully low precision
            B = Matrix{high}(inv(J))   # invert in low precision, then convert once
            upd = true
        elseif corr_id ==  3
            Id = Matrix{low}(I, 2*N, 2*N)
            B = Id + 
            upd = true
        else 
            Id = Matrix{low}(I, 2*N, 2*N)
            
            x1 = low.(u0[1:N])        # convert to low precision
            x2 = low.(u0[N+1:end])    # convert to low precision
            gL = low(9.81)
            
            A11 = zeros(low, N, N)
            A12 = Dx_low                          # already low precision
            A21 = -Dx_low * Diagonal((x2.^2) ./ (x1.^2) .- gL .* x1)
            A22 = 2 * Dx_low * Diagonal(x2 ./ x1)
            
            A = [A11  A12;
                 A21  A22]
            
            B = Id + (low(dt) / 2) * A + (low(dt.^2) / 4) * (A*A)  # now J is fully low precision
            #F = svd(J)                        # computed in Float32
            #U, S, V = low.(F.U), low.(F.S), low.(F.V)  # Artifitial 
            #B = Matrix{high}(V * Diagonal(one(low) ./ S) * U')
            upd = true
        end
    else 
        B = Id_high
        upd = false
    end


    result = @timed begin
        temp_u = copy(u0)
        for _ in 1:nt
            temp_u = implicit_midpoint_step(temp_u, Dx_high, Dx_low, high(dt), high, low, correct = correct,n_corr= n_corr, B = B, update = upd)
        end
        temp_u
    end

    runtime = result.time # Convert nanoseconds to seconds
    u = result.value  # Final value after all steps

    # L2 Error
    norm_L2(u, dx) = sqrt(sum(abs2, u) * dx)
    error_L2 = norm_L2(uref - u, dx)

    return (
        error = error_L2,
        time = runtime,
        u = u,
        uref = uref,
        x = x,
    )
    

end


```
